```python
main_path=os.path.join('Tensorflow', 'workspace', 'images')

img_dir = os.path.join(main_path,'collectedimages')
liste_image= glob.glob(os.path.join(img_dir, '*.jpeg'))
liste_label= glob.glob(os.path.join(img_dir, '*.xml'))
if not os.path.exists(os.path.join(main_path,'test')) and not os.path.exists(os.path.join(main_path,'train')):
    os.mkdir(os.path.join(main_path,'test'))
    os.mkdir(os.path.join(main_path,'train'))
pourcentage=int(input('entrer le pourcentage des images qui seront depalace dans train:'))
while pourcentage>99 or pourcentage<95:
    pourcentage=int(input('erreur!! reentrer le pourcentage des images qui seront depalace dans train:'))

for path in ['train','test']:
    if path=='train':
        nb=int(((len(liste_image))*pourcentage)/100)
    else:
        liste_image=glob.glob(os.path.join(img_dir, '*.jpeg'))
        nb=len(liste_image)
    print(nb)
    for i in range(nb):
        print(liste_image[i],'----',i)
        d+=1
        img=os.path.split(liste_image[i])[1].split('.')[0]
        for j in range(len(liste_label)):
            label=os.path.split(liste_label[j])[1].split('.')[0]
            if img==label:
                shutil.move(liste_label[j],os.path.join(main_path,path,os.path.split(liste_label[j])[1]))
                print(img,'==',label)
                print(liste_label[j],'||',os.path.join(main_path,path,os.path.split(liste_label[j])[1]))
                break
        print(liste_image[i],'||',os.path.join(main_path,path,os.path.split(liste_image[i])[1]))
        shutil.move(liste_image[i],os.path.join(main_path,path,os.path.split(liste_image[i])[1]))
print(f"{pourcentage}% seront dans la partie train et {100-pourcentage}% seront dans la partie test ")
```
